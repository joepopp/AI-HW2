```python
class WaterJugPuzzle():

    # We model the state using a list of two numbers: amount of water in the large jug, amount of water in the small jug
    def __init__(self):
        self.jugs = [0,0]


    def large_filled_improperly(self):
        if (self.jugs[0] < 0 or self.jugs[0] > 5):
            return True

    def small_filled_improperly(self):
        if (self.jugs[1] < 0 or self.jugs[1] > 3):
            return True

    def state_ok(self):
        return not (self.large_filled_improperly() or self.small_filled_improperly())

    def save_state(self):
        self._jugs  = list(self.jugs)

    def undo_state(self):
        self.jugs  = list(self._jugs)

    ####################################################################
    # Rule 1: Fill a jug.
    ####################################################################
    def R1_fill_jug(self, size):

        self.save_state()

        if (size == 'large'):
            self.jugs[0] = 5
            if not self.state_ok(): self.undo_state()

        elif (size == 'small'):
            self.jugs[1] = 3
            if not self.state_ok(): self.undo_state()

        else:
            raise ValueError('Size must be either large or small')

    ####################################################################
    # Rule 2: Empty a jug.
    ####################################################################
    def R2_empty_jug(self, size):

        self.save_state()

        if (size == 'large'):
            self.jugs[0] = 0
            if not self.state_ok(): self.undo_state()

        elif (size == 'small'):
            self.jugs[1] = 0
            if not self.state_ok(): self.undo_state()

        else:
            raise ValueError('Size must be either large or small')

    ####################################################################
    # Rule 3: Pour water from the large to the small jug
    # We check to see whether the large jug can empty its contents fully
    # into the small jug or the small jug would overfill and pour accordingly
    # If this transition results in an invalid state we undo the state-change.
    ####################################################################
    def R3_large_to_small(self):

        self.save_state()

        if (self.jugs[0] > (3 - self.jugs[1])):
            self.jugs[0] -= (3 - self.jugs[1])
            self.jugs[1] = 3
            if not self.state_ok(): self.undo_state()

        elif (self.jugs[0] <= (3 - self.jugs[1])):
            self.jugs[1] += self.jugs[0]
            self.jugs[0] = 0
            if not self.state_ok(): self.undo_state()

    ####################################################################
    # Rule 4: Pour water from the small to the large jug
    ####################################################################
    def R4_small_to_large(self):

        self.save_state()

        if (self.jugs[1] > (5 - self.jugs[0])):
            self.jugs[1] -= (5 - self.jugs[0])
            self.jugs[0] = 5
            if not self.state_ok(): self.undo_state()

        elif (self.jugs[1] <= (5 - self.jugs[0])):
            self.jugs[0] += self.jugs[1]
            self.jugs[1] = 0
            if not self.state_ok(): self.undo_state()

    #########################################################################
    # We need two invariants to ensure the state of the system is consistent
    #########################################################################
    def large_filled_properly(self):
        return not self.large_filled_improperly()

    def small_filled_properly(self):
        return not self.small_filled_improperly()

    ####################################################################
    # Display amount of water in each jug
    ####################################################################
    def print_state(self):
        print("===> large: {l}, small: {s}".format(l=self.jugs[0], s=self.jugs[1]))
```

```python
#@title Forming the Solution

mcf = WaterJugPuzzle()

# print the initial state
mcf.print_state()


print("-----------------------")
# apply a rule and print the resulting state
mcf.R1_fill_jug('large')
mcf.print_state()

print("-----------------------")
mcf.R3_large_to_small()
mcf.print_state()

print("-----------------------")
mcf.R2_empty_jug('small')
mcf.print_state()

print("-----------------------")
mcf.R3_large_to_small()
mcf.print_state()

print("-----------------------")
mcf.R1_fill_jug('large')
mcf.print_state()

print("-----------------------")
mcf.R3_large_to_small()
mcf.print_state()
```
